### Срезы

`list[start : stop : step]`

, где:  
* start — индекс начала среза (включительно). По умолчанию равен 0.
* stop  — индекс окончания среза (исключительно). По умолчанию равен len(list) - 1.
* step  — шаг. По умолчанию равен 1

Минусы: в памяти создается копия (срез)

```python
li = [1, 2, 3, 4]   # индексы: [0, 1, 2, 3]

li[1:3]     # => [2, 3]
li[2:]      # => [3, 4]
li[:3]      # => [1, 2, 3]
li[::2]     # => [1, 3]
li[::-1]    # => [4, 3, 2, 1]

'''Особенность'''
print(li[1:3])    # Срезы никогда не падают с выходом за границы
print(li[1:3])
li[1:3]
li[1:3]
```

#### Как думаете, какой будет результат?

```python
li[0:100]
```

```python
print('Привет мир!'[7:10])    # => 'мир'
print('Привет мир!'[0:6:2])   # => 'Пие'
```

### Словарь (dict)

> Словарь - это структура данных, которая позволяет хранить коллекции пар «ключ — значение». Ключ в словарях уникальны, значения же могут дублироваться

#### Объявление

```python
{}
dict()
{'first': 1, 'second': {'inner': True}}
```

#### Методы

```python
d = {'a': 1, 'b': 2}

len(d)               # => 2
3 in d               # => False

d["a"]               # => 1
d["x"]               # => ошибка KeyError
d.get("x")           # => None
d.get("x", 0)        # => 0 (значение по умолчанию)

d.update({"c": 3})   # добавить новый ключ-значение или изменить значение

d.pop("a")           # удаляет ключ-значение и возвращает
d.pop("x", None)     # удялить без ошибки
d.popitem()          # удаляет последний добавленный ключ
d.clear()            # очищает весь словарь

d.keys()             # ключи
d.values()           # значения
d.items()            # (ключ, значение)
```

## Pattern matching (появилось в python3.10)

#### Базовая конструкция

```python
status = 200

match status:
    case 200:
        print("OK")
    case 404:
        print("Not Found")
    case 500:
        print("Server Error")
    case _:
        print("Unknown")
```

#### Базовая конструкция

```python
# data = {"name": "Иван", 'surname': "Петров"}
data = [1, 2]
data = 5


match data:
    case {"name": "Иван"}:
        print("Это Ваня")
    case n if isinstance(n, int) and n > 0:
        print("Положительное число")
    case [x, y]:
        print(f"Это список с двумя элементами: {x} и {y}")
    case _:
        print("Unknown")
```

## Циклы

#### for-in

```python
animals = ["собака", "кошка", "мышь"]

for animal in animals:
    print(f"{animal} — это млекопитающее")
```

#### range()

> Функция используется для генерации последовательностей чисел

```python
for i in range(len(animals)):
    print(f"{animals[i]} — это млекопитающее - {i}")
```

#### enumerate()

> Вместе со значением итерируемого объекта возвращает порядковый номер

```python
for i, animal in enumerate(animals):
    print(f"{i}. {animal} — это млекопитающее")
# animals[2:]
```

#### while

```python
# Продолжается до тех пор, пока указанное условие не станет ложным

i = 0
while i < len(animals):
    animal = animals[i]
    print(f"{i}: {animal} — это млекопитающее")
    i += 1
```

#### Управление потоком цикла

##### continue

```python
for i, animal in enumerate(animals):
    if i == 1:
        continue
    print(f"{i}: {animal} — это млекопитающее")
```

##### break

```python
for i, animal in enumerate(animals):
    if i == 1:
        break
    print(f"{i}: {animal} — это млекопитающее")
```

#### Что будет, если изменить объект, по которому проходимся циклом?

```python
animals = ["собака", "кошка", "мышь"]

for i, animal in enumerate(animals):
    print(f'{i} - Массив: {animals}')
    animals.append('name')

    if i == 15:  # без этого блока может выполняться бесконечно
        break
```

#### Python имеет функции первого класса

> Функция в Python это тоже объект

```python
# то же самое

print( create_adder(10)(3) )
```

##### Для каррирования лучше использовать functools.partial

> Каррирование - преобразование функции от многих аргументов в набор вложенных функций, каждая из которых является функцией

```python
from functools import partial

def create_adder(x, y):
    return x + y

add_10 = partial(create_adder, 10)
print( add_10(3) )
```

#### *args и **kwargs

```python
# Вы можете определить функцию, принимающую переменное число аргументов
def function_with_args(qwe, *args, a=None, b=None, **kwargs):
    print(f'{qwe=}')
    print(f'{a=}')
    print(f'{b=}')
    print(type(args))
    print(f'{args=}')
    print(type(kwargs))
    print(f'{kwargs=}')

function_with_args(1, 2, a='a', new_value=123)  # => (1, 2, (3, 4))
```

```python
# Именованные аргументы
def function_with_kwargs(*args, **kwargs):
    print(type(kwargs))
    print(f'{args=}')
    print(f'{kwargs=}')

function_with_kwargs(1, big="foot", loch="ness")
```

#### Область видимости

```python
x = 5

def use_local(num):
    # Локальная переменная x — это Не то же самое, что х за пределами функции
    x = num
    print(x)

def use_global(num):
    global x
    print(x)
    x = num
    print(x)
```

```python
# use_local(10)
use_global(6)
```

#### Переменные экземпляра

```python
class CustomUser:
    type='Employer'
    
    def __init__(self, name, surname):
        self.name = name
        self.surname = surname

    def get_fio(self):
        return f'{self.name} {self.surname}'
    

a = CustomUser('Петр', 'Фролов')
a.name
```

```python
a.type
```

```python
a.name = 'dfgs'
a.name
```

```python
a.get_fio()
```

#### Переменные класса

```python
class MyClass:
    @classmethod
    def get_class_name(cls):
        return cls.__name__
```

```python
MyClass.get_class_name()
```

#### Переопределение методов и добавление своих

```python
class User:
    def __init__(self, name, surname):
        self.name = name
        self.surname = surname

    def __str__(self):
        return f'Имя Фамилия: {self.name} {self.surname}'

    def get_name(self):
        return self.name
```

```python
user1 = User('Николай', 'Иванов')
```

```python
user1.get_name()
```

```python
str(user1)
```

#### Утиная типизация (Duck typing)

> Если это выглядит как утка, плавает как утка и крякает как утка, то это, вероятно, и есть утка.

```python
class MyInt:
    def __eq__(self, other_value):
        return 'gnhfjhsdjs'

class MyInt_2:
    def __eq__(self, other_value):
        return 'TRUE'

a = MyInt()
b = MyInt_2()
print(type(a))
print(type(b))
a == b
```

## Наследование

```python
class Deffault:
    pass

print(Deffault.__mro__)

class Class1(Deffault):
    pass

class Deffault(Class1):
    pass

class Class2(Deffault):
    pass

# этот класс наследует сразу от двух родительских классов: Class1 и Class2
class MultiDerived(Class2, Class1):
    pass
```

```python
MultiDerived.__mro__
```
